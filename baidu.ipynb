```python
# encoding:utf-8

import requests
import base64

'''
黑白图像上色
'''

request_url = "https://aip.baidubce.com/rest/2.0/image-process/v1/colourize"
# 二进制方式打开图片文件
f = open('10386559_121919534164_2.jpg', 'rb')
img = base64.b64encode(f.read())
f.close()
params = {"image":img}
#access_token = '25.938db687b4611796898626d16fb20694.315360000.1911831254.282335-21776638'
request_url = request_url + "?access_token=" + access_token
headers = {'content-type': 'application/x-www-form-urlencoded'}
response = requests.post(request_url, data=params, headers=headers)
if response:
    print (response.json())
with open('test_out.jpg','wb') as g:
    g.write(base64.b64decode(response.json()['image']))
    g.close()
```
